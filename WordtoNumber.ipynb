## 建立字典

```python
# -*- coding: utf-8 -*-
import jieba
jieba.set_dictionary('dict.txt.big')##加入繁體中文

import re
from collections import Counter

words = []
dict = {}
weight = []

with open('hw1-dataset.txt',encoding="utf-8") as file:
    for i,line in enumerate(file.readlines()):
        line = re.sub("[A-Za-z0-9\：\?\(\.\·\—\，\。\“ \”]", "", line)

        #建立分詞
        jieba.add_word('為什麼')
        jieba.add_word('有沒有')
        tags = jieba.lcut(line,cut_all=False)
        for tag in tags:
            if tag == "" or tag == '\t' or tag == '\n':
                continue;
            words.append(tag)
        
        c = Counter()
        index = 1
        for x in words:
            if len(x)>1 and x != '\r\n':
                c[x] += 1
                dict.setdefault(x, index)
                index = index + 1

## 建立字換編號對應表

with open('WordNumber.txt', 'a') as WordNum:
    for key,values in  dict.items():
        text = key + " " + str(values) + '\n'
        WordNum.write(text)
```
